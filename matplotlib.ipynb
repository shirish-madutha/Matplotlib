```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.randint(1,100,5)
y=np.random.randint(1,100,5)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,lw="5",ls="--",c="indigo",marker="o", ms=20, mec="hotpink" ,mfc="#ACA39A")
plt.show()
```

```python
font1={"c":"green","size":10}
font2={"c":"indigo","size":15}
plt.xlabel("x-axis",fontdict=font1)
plt.ylabel("y-axis",fontdict=font1)
plt.title("Graph", fontdict=font2, loc="left")
plt.plot(x,y)
plt.grid(axis="both", ls=":", lw=1, c='g')
plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=20)
y = np.random.randint(100, size=20)
colors = np.random.randint(100, size=20)
sizes = 10 * np.random.randint(100, size=20)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)

plt.show()
```

```python
# bar plot
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width=0.5, color="INDIGO")
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height=0.5, color="GREY")
plt.show()
```

```python
# histogram
arr=np.array([5,2,1,5,1,2,4,2,3,8,5,6,4,9])
plt.hist(arr)
plt.show()
```

```python
# pie chart
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "BlueBerry", "Grapes", "Orange"]
colors=["red", "blue", "green", "orange"]

plt.pie(y, labels = mylabels, colors=colors)
plt.legend(title = "Four Fruits:", loc="upper right")
plt.show() 
```
